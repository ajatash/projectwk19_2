```python
import pandas as pd
import numpy as np
```

```python
file = "CountiesData_noGeometry_cleaned.csv"
file_df = pd.read_csv(file)
file_df
```

```python
top_ten = file_df.sort_values('Population', ascending = False).head(10)
top_ten
```

```python
file_df.head()
```

```python
file_df['Updated_name'] = file_df["Name"].str[:-18]
file_df
```

```python
#drop old name

dropped = file_df.drop(columns=['Name'])
dropped.head()
```

```python
columns2 = ["ID", "Updated_name", "Population", "TotalAgeOvr65"]
new_counties = dropped[columns2].copy()

new_counties.head()
```

```python
updated_county = new_counties.assign(Other_column = new_counties['Population'] - new_counties['TotalAgeOvr65'], Other = lambda x: np.log(x.Other_column) )
updated_county
```

```python
columns3 = ["ID", "Updated_name", "Population", "TotalAgeOvr65", "Other_column"]
new_counties = updated_county[columns3].copy()
new_counties.head()
```

```python
final_county = new_counties.sort_values('Population', ascending=False)
final_county
```

```python
#export cleaned dataframe to csv

final_county.to_csv('Resources/cleaned_counties.csv')
```

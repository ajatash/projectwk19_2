```python
import pandas as pd
```

```python
asthma_er = "asthma_er.csv"
asthma_er = pd.read_csv(asthma_er)
asthma_er
```

```python
asthma_er_clean = asthma_er.dropna()
asthma_er_clean
```

```python
asthma_er_clean = asthma_er_clean.rename(
    columns={"fips": "County"})
asthma_er_clean.head()
```

```python
asthma_hosp = "asthma_hosp.csv"
asthma_hosp = pd.read_csv(asthma_hosp)
asthma_hosp
```

```python
asthma_hosp_clean = asthma_hosp.dropna()
asthma_hosp_clean
```

```python
asthma_hosp_clean.columns
```

```python
asthma_hosp_clean = asthma_hosp_clean.rename(
    columns={"fips": "County"})
asthma_hosp_clean.head()
```

```python
merge_asthma = pd.merge(asthma_hosp_clean, asthma_er_clean, on="County")
merge_asthma
```
